```python
GET :
    GET method send data through URL.
    This method is limited to a maximum number of characters.
    This method is limited in length.
    It can be visible for everyone that's why it is not secure.
    Example :
        Any Google search.
        
POST :
    POST method send data through HTTP body, not URL.
    It has no limitation.
    It is more secure than GET.
    Example :
        Email_ID, Password.
```
